```python
import requests

url = 'https://www.nationmaster.com/country-info/stats/Geography/Area/Land'
html = requests.get(url).content
html
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, "lxml")  # Check documentation to specify 'features'
soup # lxml = different tagging 
```

```python
table = soup.find_all('table')[0]#, attrs = {'class':'table table-active table-main table-paginated table-sortable resptable tablesorter tablesorter-default'})


```

```python
#country  = [x.text for x in table.find_all('td', attrs = {'class':'full'})]

country = [x.find('span', attrs = {'class': 'full'}) for x in table.find_all('td')]
```
